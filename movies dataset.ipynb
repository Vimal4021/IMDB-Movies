```python
#creaing one more column 

def rating(num):
    if num>7 and num<10:
        return "Excellent "
    elif num>4 and num<7:
        return "good movie"
    else:
        return "avg movie"
```

```python
df['Movies Review']=df['IMDB_Rating'].apply(rating)
```
